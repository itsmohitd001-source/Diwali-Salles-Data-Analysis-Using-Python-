***Exploratory data analysis***

```python
a=sns.countplot(x="Gender",data=df)
for container in a.containers:
    a.bar_label(container)

plt.show()
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
x=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=x)
```

>>>females are the majority buyers


age group


```python
a=sns.countplot(x="Age Group",data=df, hue="Gender")
for container in a.containers:
    a.bar_label(container)
```

```python
x=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=x)
```

 26-35 is the age group where most of the shopping happens!

```python
x=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="State",y="Orders",data=x)
```

```python
x=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="State",y="Amount",data=x)
```

```python
a=sns.countplot(x="Marital_Status",data=df,)
sns.set(rc={"figure.figsize":(5,2)})
for container in a.containers:
    a.bar_label(container)
```

```python
x=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(3,3)})
sns.barplot(x="Marital_Status",y="Amount",data=x, hue="Gender")
```

married women are the highest number of buyers.


Occupation

```python
a=sns.countplot(x="Occupation",data=df,)
sns.set(rc={"figure.figsize":(20,7)})
for container in a.containers:
    a.bar_label(container)
```

```python
x=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,7)})
sns.barplot(x="Occupation",y="Amount",data=x, )
```

```python
a=sns.countplot(x="Product_Category",data=df)
sns.set(rc={"figure.figsize":(25,10)})
for container in a.containers:
    a.bar_label(container)
```

```python
x=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,10)})
sns.barplot(x="Product_Category",y="Amount",data=x)
```

food electronics and clothing are the most purchased products.

```python
x=df.groupby(["Product_ID"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,7)})
sns.barplot(x="Product_ID",y="Amount",data=x)
```
